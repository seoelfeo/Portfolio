### 문제 2
접속 이력 테이블 (표1) 을 참고하여 표 3과 같은 결과를 도출하는 쿼리 작성
- 표2 비고 내용을 확인
- 쿼리 결과 정렬 필수
- 첨부파일 : 문제2.tsv


```python
%%sql
SELECT *, ROW_NUMBER() OVER(PARTITION BY [ID] ORDER BY [DT])
FROM 문제2
```

```python
%%sql
WITH t as (
SELECT *, ROW_NUMBER() OVER(PARTITION BY [ID] ORDER BY [DT]) as i
FROM 문제2)

SELECT ID, min(DT) as START_DT, max(DT) as END_DT
FROM t
GROUP BY DATEDIFF(day, DT, i)
```

### 문제 3-2
호텔 검색 로그 테이블과 호텔 예약 로그 테이블을 이용하여 아래 결과 도출
쿼리 작성
- 호텔 예약이 없었던 유저들의 호텔 검색 횟수와 검색한 호텔 수 추출
- 표 4 에서 스키마 정보 확인

```python
%%sql
SELECT s.USER_ID, COUNT(s.USER_ID) as SEARCH_CNT, COUNT(DISTINCT(s.HOTEL_CD))
FROM 문제3_search s LEFT JOIN 문제3_resv r ON r.USER_ID=s.USER_ID
WHERE r.USER_ID IS NULL
```

```python
%%sql
SELECT s.USER_ID, COUNT(*) as SEARCH_CNT, COUNT(DISTINCT(s.HOTEL_CD)) as SEARCH_HOTEL_CNT
FROM 문제3_search s LEFT JOIN 문제3_resv r ON r.USER_ID=s.USER_ID
WHERE r.USER_ID IS NULL
GROUP BY s.USER_ID
```

### 문제 3-3
호텔 검색 로그 테이블과 호텔 예약 로그 테이블을 이용하여 아래 결과 도출
쿼리 작성
- 각 유저 별 호텔 검색 수, 검색한 호텔 수, 예약한 호텔 수 추출
- 표 5 에서 스키마 정보 확인

```python
%%sql
SELECT *
FROM 문제3_search 
UNION 
SELECT *
FROM 문제3_resv
```

```python
%%sql
with t as (
SELECT *
FROM 문제3_search 
UNION 
SELECT *
FROM 문제3_resv)

SELECT 
    USER_ID,
    sum(CASE WHEN ACTION = '검색' THEN 1 ELSE 0 END) as SEARCH_CNT,
    COUNT(DISTINCT CASE WHEN ACTION = '검색' THEN HOTEL_CD END) as SEARCH_HOTEL_CNT,
    COUNT(DISTINCT CASE WHEN ACTION = '예약' THEN HOTEL_CD END) as RESERVE_HOTEL_CNT
FROM t
GROUP BY USER_ID
```
