### 문제 1
매출 테이블 (표 1 스키마 정보 확인) 을 참고하여 표 3과 같은 결과를 도출하는
쿼리 작성
- 표 2 결과 테이블 스키마 정보 비고 내용을 확인
- 쿼리 작성시 각 컬럼의 데이터 타입 정의 필수
- 쿼리 결과 정렬 필수
- 첨부파일 : 문제1.tsv

```python
%%sql
SELECT *, DATE_ADD(FST_CRE_UTC_DTTM, INTERVAL 9 HOUR) as BASIS_DT
FROM 문제1
ORDER BY 8, 5
LIMIT 5
```

```python
%%sql
SELECT 
    CAST(DATE_ADD(FST_CRE_UTC_DTTM, INTERVAL 9 HOUR) AS DATE) as BASIS_DT, 
    SVC_TYPE_NM,
    COUNT(DISTINCT(ORDER_ID)) as ORDER_CNT,
    COUNT(DISTINCT(RESV_ID)) as RESV_CNT,
    COUNT(DISTINCT(USER_ID)) as USER_CNT,
    SUM(PMT_AMT) as MPT_AMT,
    SUM(CASE WHEN PG_DEAL_METHOD_TYPE = 'CREDIT' THEN PMT_AMT ELSE 0 END) as CREDIT_PMT_AMT,
    SUM(CASE WHEN PG_DEAL_METHOD_TYPE = 'NAVER_PAY' THEN PMT_AMT ELSE 0 END) as NAVER_PMT_AMT
FROM 문제1
GROUP BY 1, 2
ORDER BY 1
```

```python
%%sql
SELECT DATE(FST_CRE_UTC_DTTM) as BASIS_DT, SVC_TYPE_NM, COUNT(DISTINCT(ORDER_ID)) as ORDER_CNT
FROM 문제1
WHERE DATE(FST_CRE_UTC_DTTM) > '2022-05-20'
GROUP BY BASIS_DT, SVC_TYPE_NM
ORDER BY BASIS_DT asc
```

### 문제 2
접속 이력 테이블 (표1) 을 참고하여 표 3과 같은 결과를 도출하는 쿼리 작성
- 표2 비고 내용을 확인
- 쿼리 결과 정렬 필수
- 첨부파일 : 문제2.tsv


```python
%%sql
SELECT
    ID,
    ROW_NUMBER() OVER (PARTITION BY ID ORDER BY DT) AS rn,
    DATEDIFF('2022-02-01', DT) AS grp,
    DT
FROM 문제2
```

```python
%%sql
WITH g as (
SELECT
  ID,
  ROW_NUMBER() OVER (PARTITION BY ID ORDER BY DT) AS rn,
  DATEDIFF('2022-02-01', DT) AS grp,
  DT
FROM 문제2)

SELECT
  COUNT(*) AS consecutiveDates,
  MIN(DT) AS minDate,
  MAX(DT) AS maxDate
FROM g
GROUP BY grp
ORDER BY 1 DESC, 2 DESC
```

```python
%%sql
SELECT 
    ID,
    DT,
    DATEDIFF('2022-04-01', DT) - ROW_NUMBER() OVER(PARTITION BY ID ORDER BY DT)
FROM 문제2
```

```python
%%sql
WITH t as (
SELECT 
    ID,
    DT,
    DATEDIFF('2022-04-01', DT) - ROW_NUMBER() OVER(PARTITION BY ID ORDER BY DT) as i
FROM 문제2)

SELECT ID, min(DT) as START_DT, max(DT) as END_DT, COUNT(*) CNT
FROM t
GROUP BY DATEDIFF(DT, i), ID
```

### 문제 3-2
호텔 검색 로그 테이블과 호텔 예약 로그 테이블을 이용하여 아래 결과 도출
쿼리 작성
- 호텔 예약이 없었던 유저들의 호텔 검색 횟수와 검색한 호텔 수 추출
- 표 4 에서 스키마 정보 확인

```python
%%sql
SELECT s.USER_ID, COUNT(*) as SEARCH_CNT, COUNT(DISTINCT(s.HOTEL_CD)) as SEARCH_HOTEL_CNT
FROM 문제3_search s LEFT JOIN 문제3_resv r ON r.USER_ID=s.USER_ID
WHERE r.USER_ID IS NULL
GROUP BY s.USER_ID
```

### 문제 3-3
호텔 검색 로그 테이블과 호텔 예약 로그 테이블을 이용하여 아래 결과 도출
쿼리 작성
- 각 유저 별 호텔 검색 수, 검색한 호텔 수, 예약한 호텔 수 추출
- 표 5 에서 스키마 정보 확인

```python
%%sql
with t as (
SELECT *
FROM 문제3_search 
UNION 
SELECT *
FROM 문제3_resv)

SELECT 
    USER_ID,
    sum(CASE WHEN ACTION = '검색' THEN 1 ELSE 0 END) as SEARCH_CNT,
    COUNT(DISTINCT CASE WHEN ACTION = '검색' THEN HOTEL_CD END) as SEARCH_HOTEL_CNT,
    COUNT(DISTINCT CASE WHEN ACTION = '예약' THEN HOTEL_CD END) as RESERVE_HOTEL_CNT
FROM t
GROUP BY USER_ID
```
